```python
df = pd.read_csv('C:/Users/Jefrin Raj/Downloads/Telegram Desktop/train.csv')
```

```python
df.head()
```

### HISTOGRAM FOR NUMERICAL COLUMN

```python
numeric_cols=["Age","SibSp"]
for col in numeric_cols[:5]:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col].dropna(),kde=True, color="blue")
    plt.title(f"distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("frequency")
    plt.tight_layout()
    plt.show()
```

### COUNT PLOT FOR CATEGIORICAL COLUMN

```python

import seaborn as sns
import matplotlib.pyplot as plt

categorical_col = ['Sex','Cabin']
plt.figure(figsize=(3, 2))
for col in categorical_col:
    sns.countplot(x=df[col], color="blue")
    plt.title(f" Count Plot of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

### BOX PLOTS

```python
numeric_cols = ["Age", "Fare"]

for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x="Survived", y=col, data=df, color="skyblue")
    plt.title(f"Survived vs {col}")
    plt.xlabel("Survived")
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```

```python
#correlation heatmap
selected_cols = ["Age", "Fare", "Survived", "Pclass", "SibSp", "Parch"]
corr = df[selected_cols].corr()
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
